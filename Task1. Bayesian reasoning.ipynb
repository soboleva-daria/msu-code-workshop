## Вероятностная модель посещаемости курса

Рассматривается модель посещаемости студентами ВУЗа одной лекции по курсу.
Аудитория данного курса состоит из студентов профильного факультета ($a$ студентов), каждый из которых посещает лекцию с 
вероятностью $p_1$
и студентов других факультетов ($b$ студентов), каждый из которых посещает лекцию с вероятностью $p_2$.


Величины $a$ и $b$ имеют дискретное равномерное распределение в качестве априорного: $$a \sim Unif[a_{min}, a_{max}],$$ 
$$b \sim Unif[b_{min}, b_{max}]$$

Пусть лекцию посетило $c$ студентов. Тогда случайная величина $c|a, b \sim Bin(a, p_1) + Bin(b, p_2)$.
Пусть далее на лекции по курсу ведется запись студентов. При этом каждый студент записывается сам,
а также, быть может, записывает своего товарища (с вероятностью $p_3$), которого на лекции на самом деле нет. 
Пусть $d$ -- это общее количество записавшихся на данной лекции. Тогда $d|c \sim c + Bin(c, p_3)$.

Далее модель, рассмотренную выше, будем обозначать $(1)$.
Модель $(2)$ представляет собой упрощенную версию модели $(1)$. Для нее вводится распределение
случайной величины $c|a, b \sim Poiss(ap_1 + bp_2)$.

Получившаяся модель обладает также условием $p(a, b, c, d) = p(d|c)p(c|a, b)p(a)p(b)$.

### $1$. Нахождение математических ожиданий и дисперсий для распределений $p(a), p(b), p(c), p(d)$

Все результаты, представленные в данном пункте, получены теоритически. Следует отметить, что они полностью совпали с результатами,

Для вывода большинства формул будем пользоваться законом о полном математическом ожидании:
    
Пусть $B_1 \dots B_n$ -- полная группа событий ($B_1 \dots B_n$ -- суть попарно несовместные, имеющие
положительные вероятности и $\cup_{i=1}^{n} B_i = \Omega$).
$F(x|B_1) \dots F(x|B_n)$ -- условные функции распределения величины $X$.

Воспользуемся формулой полной вероятности для функции распределения $F(x)$ случайной величины $X$: $F(x) = \sum_{i=1}^n P(B_i) F(x|B_i)$.
    
А следовательно, по определению математического ожидания, имеем:
$\mathbb{E} (X) = \sum_{i=1}^n P(B_i) \mathbb{E}(x|B_i)$.

$a \sim Unif[a_{min}, a_{max}]$

$\mathbb{E} (a) = \frac{1}{a_{max} - a_{min} + 1}\sum_{a_{min}}^{a_{max}} a = \frac{a_{max} + a_{min}}{2} = 82.5$

$\mathbb{E} (a^2) = \frac{1}{a_{max} - a_{min} + 1}\sum_{a_{min}}^{a_{max}} a^2 = \frac{1}{6} (2 a_{min}^2 + 
2 a_{min} a_{max} - a_{min} + 2 a_{max}^2 + a_{max})$

$\mathbb{D} (a) = \mathbb{E} (a^2) - (\mathbb{E} (a))^2 = \frac{(a_{max} - a_{min} + 1)^2 - 1}{12} = 21.25$

    
$b \sim Unif[b_{min}, b_{max}]$

$\mathbb{E} (b) = \frac{b_{max} + b_{min}}{2} = 550$

$\mathbb{D} (b) = \frac{(b_{max} - b_{min} + 1)^2 - 1}{12} = 850$
    
    
$c|a, b \sim Bin(a, p_1) + Bin(b, p_2)$

$\mathbb{E} (c) = \frac{1}{a_{max} - a_{min} + 1} \frac{1}{b_{max} - b_{min} + 1} \sum_{a_{min}}^{a_{max}} \sum_{b_{min}}^{b_{max}}\mathbb{E} (c|a, b) = $ 
$= \{\mathbb{E} (c|a, b) = \mathbb{E} (Bin(a, p_1) + Bin(b, p_2)) = \mathbb{E} (Bin(a, p_1)) + \mathbb{E} (Bin(b, p_2)) = ap_1 + bp_2\}$

$ = \frac{1}{a_{max} - a_{min} + 1} \frac{1}{b_{max} - b_{min} + 1} \sum_{a_{min}}^{a_{max}} \sum_{b_{min}}^{b_{max}} ap_1 + bp_2$
$ = p_1\mathbb{E} (a) + p_2 \mathbb{E} (b) = 13.75$

$\mathbb{E} (c^2) = \frac{1}{a_{max} - a_{min} + 1} \frac{1}{b_{max} - b_{min} + 1} \sum_{a_{min}}^{a_{max}} \sum_{b_{min}}^{b_{max}} ap_1 (1 - p_1) + a^2p_1^2 + bp_2 (1 - p_2) + b^2p_2^2 + 2ap_1p_2$

$\mathbb{D} (c) = p_1 (1 - p_1)\mathbb{E} (a) + p_2 (1 - p_2)\mathbb{E} (b) + p_2^2\mathbb{D} (b) + p_1^2\mathbb{D} (a) = 13.17$
    
    
$c|a, b \sim Poiss(ap_1 + bp_2)$

$ \mathbb{E} (c) =  \frac{1}{a_{max} - a_{min} + 1} \frac{1}{b_{max} - b_{min} + 1}\sum_{a_{min}}^{a_{max}} \sum_{b_{min}}^{b_{max}} ap_1 + bp_2$
$ = p_1\mathbb{E} (a) + p_2 \mathbb{E} (b) = 13.75$

$\mathbb{E} (c^2) = \frac{1}{a_{max} - a_{min} + 1} \frac{1}{b_{max} - b_{min} + 1} \sum_{a_{min}}^{a_{max}} \sum_{b_{min}}^{b_{max}} ap_1 + a^2p_1^2 + bp_2 + b^2p_2^2 + 2ap_1p_2$

$\mathbb{D} (c) = p_1\mathbb{E} (a) + p_2\mathbb{E} (b) + p_2^2\mathbb{D} (b) + p_1^2\mathbb{D} (a) = 14.05$
    
    
$d|c \sim c + Bin(c, p_3)$

$\mathbb{E} (d) = \sum_{c} \mathbb{E} (d|c) p(c) = \sum_{c} (p_1 \mathbb{E} (a) + p_2 \mathbb{E} (b) + cp_3) p(c)$

### $2$. Уточнение прогноза для величины $b$ по мере прихода новой косвенной информации

Имеем $\mathbb{E}(a) = 82, \mathbb{E}(d) = 18$.

Вычислим математическое ожидание и дисперсию для распределений $p(b), p(b|a = 82), p(b|d=18), p(b|a=82, d=18)$.

Результаты для $p(b)$ представлены выше


$p(b|a) = \frac{p(b, a)}{p(a)} = \frac{p(b)p(a)}{p(a)} = p(b)$

$\mathbb{E}(b|a = 82) = \mathbb{E}(b) = \frac{b_{max} + b_{min}}{2} = 550$  

$\mathbb{D}(b|a = 82) = \mathbb{D}(b) = \frac{(b_{max} - b_{min} + 1)^2 - 1}{12} = 850$

    
$p(b|d) = \frac{p(d|b)p(b)}{\sum_{b}p(d|b)p(b)} = \frac{\sum_{a}\sum_{c}p(d|c)p(c|a, b)}{\sum_{a}\sum_{c}\sum_{b}p(d|c)p(c|a, b)}$

$\mathbb{E} (b|d) = \sum_{b} b p(b|d)$

$\mathbb{E} (b|d)^2 = \sum_{b} b^2 p(b|d)$

$\mathbb{D} (b|d) = \sum_{b} b^2 p(b|d) - (\sum_{b} b p(b|d))^2$

Модель (1) $\mathbb{E} (b|d = 18) = 550.07, \mathbb{D} (b|d = 18) = 848.04.$ 

Модель (2) $\mathbb{E} (b|d = 18) = 550.1, \mathbb{D} (b|d = 18) = 848.13.$


$p(b|a, d) = \frac{p(a, b, d)}{p(d, a)} = \frac{\sum_{c} p(a, b, d|c)p(c)}{\sum_{b}\sum_{c}p(d|c)p(c|a, b)p(b)p(a)} = 
  = \frac{\sum_{c}p(d|c)p(c|a, b)}{\sum_{b}\sum_{c}p(d|c)p(c|a, b)}
    $
    
$\mathbb{E} (b|d, a) = \sum_{b} b p(b|d, a)$

$\mathbb{E} (b|d, a)^2 = \sum_{b} b^2 p(b|d, a)$

$\mathbb{D} (b|d, a) = \sum_{b} b^2 p(b|d, a) - (\sum_{b} b p(b|d, a))^2$

Модель (1) $\mathbb{E} (b|a = 82, d = 18) = 550.09, \mathbb{D} (b|a = 82, d = 18) = 848.$

Модель (2) $\mathbb{E} (b|a = 82, d = 18) = 550.12,  \mathbb{D} (b|a = 82, d = 18) = 848.09.$

### $3$. Изменение относительной важности параметров $a, b$ для оценки $c$ при изменении параметров модели


Имеем $\mathbb{E}(a) = 82, \mathbb{E}(b) = 550$

Выразим $\mathbb{D}(c| a)$ и $\mathbb{D}(c| b)$ как функции $p_1, p_2$, воспользовавшись вычислениями пункта $1$:
    
модель (1)

$\mathbb{D}(c| a) = p_1 (1 - p_1)a + p_2 (1 - p_2)\mathbb{E} (b) + p_2^2\mathbb{D} (b)$

$\mathbb{D}(c| b) = p_1 (1 - p_1)\mathbb{E} (a) + p_2 (1 - p_2)b + p_1^2\mathbb{D} (a)$

модель (2)
        
$\mathbb{D}(c| a) = p_1 a + p_2 \mathbb{E} (b) + p_2^2\mathbb{D} (b)$

$\mathbb{D}(c| b) = p_1 \mathbb{E} (a) + p_2 b + p_1^2\mathbb{D} (a)$


Так как $a = \mathbb{E}(a), b = \mathbb{E}(b)$, то неравенство $\mathbb{D}(c|b) \lt \mathbb{D}(c|a)$ эквивалентно неравенству $p_1^2\mathbb{D}(a) \lt p_2^2 \mathbb{D}(b)$,
которое в купе с физическими ограничениями на $p_1, p_2$ определеяет множество, отделенное от своего дополнения до $(0, 1) \times (0, 1)$ прямой линией.

А значит, множества, $\{(p_1, p_2)\| \mathbb{D}(c|b) \lt \mathbb{D}(c|a)\} и 
\{(p_1, p_2)\| \mathbb{D}(c|b) \ge \mathbb{D}(c|a)\}$ 
являются линейно разделимыми для обеих моделей.

### $4$. Анализ различий

Отличие моделей (1) и (2) наблюдается в оценке параметра $b$. Если у нас есть значения величины $b$, то
различаться будут распределения $p(b|d)$ и $p(b|a, d)$. Данные изменения легко можно обьяснить отличиями
в поведении случайной величины $c$. В модели (1) она распределена как сумма двух биномиальных случайных величин (при известных $a, b$),
а в модели (2) в аналогичной ситуации она является пуассоновской случайной величиной. Так как $d$ зависит от $c$, то и распределений $p(d)$
различаются, что влияет на апостериорные распределения $p(b|d), p(b|a, d)$. Дополнительно, данные различия влияют на время вычисления
необходимых распределений.
